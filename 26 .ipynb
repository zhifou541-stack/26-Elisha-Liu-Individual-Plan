These seven variables are:

+ **Experience:** character variable. It shows the players experience level. The categories — “Amateur,” “Regular,” “Veteran,” and “Pro” — indicate increasing levels of game proficiency in that order.

+ **subscribe:** logical variable. It indicates whether a player subscribed to the game.'True' means the player did subscribe the game, 'False' means

+ **name:** character variable. It is the display name chosen by each player in the game.

+ **gender:** character variable. It shows the gender of each player.

+ **Age：** numeric variable. It shows the player's age.

Specific Question:

Can Age and played_hours predict whether a player subscribes in the players.csv dataset?


Based on what I saw earlier in the data description, it looks like played_hours might have some connection with whether a user subscribes. Players who don’t subscribe usually have 0 hours played. But relying only on played_hours doesn’t seem accurate, because some subscribed users barely play at all, and some non-subscribed users actually play for a long time. For the wrangling step, I will select the subscribe, Age, and played_hours variables, then convert subscribe from a character variable into a factor, and remove rows containing missing values in any of these three variables

3.Exploratory Data Analysis and Visualization

```r
library(tidyverse)
url1='https://drive.google.com/uc?export=download&id=1LGq5BTBDuA8hTXNmLyc1A0I6Q0ifIL6e'
players<-read_csv(url1)
head(players)
```

```r
players_clean<- players |> mutate(subscribe = as.factor(subscribe)) |> 
select(subscribe, Age, played_hours) |>
drop_na(subscribe, Age, played_hours)
head(players_clean)
```

```r
players_mean_value <- players_clean|> summarise(mean_age = mean(Age), mean_played_hours = mean(played_hours))
players_mean_value 
```

```r
options(repr.plot.width = 15, repr.plot.height = 5)
players_clean |> ggplot(aes(x = Age)) + 
 geom_histogram(binwidth = 5) +
  labs( title = "Distribution of Player Age", x = "Age (years)", y = "Count")
```

From this histogram, we can see the age distribution of the players. Most players are between 13 and 27 years old, with the largest concentration in the 13 to 18 age range.
