In the file players.csv, there are 196 observations and 7 variables in this file.

Among these variables, two are quantitative: played_hours and Age.
After calculating summary statistics, the average play time across all players is 5.42 hours, and the average player age is 19.37 years.

```r
summary_table <- data.frame(Variable = c("played_hours", "Age"), Mean = c(5.42, 19.37))
summary_table
```

These seven variables are:

+ **Experience:** character variable. It shows the players experience level. The categories — “Amateur,” “Regular,” “Veteran,” and “Pro” — indicate increasing levels of game proficiency in that order.

+ **subscribe:** logical variable. It indicates whether a player subscribed to the game.'True' means the player did subscribe the game, 'False' means 
the player did not subscribe the game.

+ **hashedEmail:** character variable. It is a hashed(anonnymized) version of the player's email.

+ **played_hours :** numeric variable. It shows the number of hours the player has spent on the game.

+ **name:** character variable. It is the display name chosen by each player in the game.

+ **gender:** character variable. It shows the gender of each player.

+ **Age：** numeric variable. It shows the player's age.

Based on the data, we can also see that subscription status seems to have some relationship with play time. Although players marked as “TRUE” don’t always play a lot — some even have just 0.1 hours — players marked as “FALSE” tend to play almost 0 hours overall.
